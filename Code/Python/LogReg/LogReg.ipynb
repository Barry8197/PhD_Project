```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, log_loss, accuracy_score, r2_score
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import sklearn as sk
```

```python
df = pd.read_csv('./ExprData/matExpr_padj.csv' , index_col=0)
```

```python
df
```

```python
X = df.iloc[: , :-1]
y = df.iloc[: , -1] - 1
```

```python
X_train, X_tmp, y_train, y_tmp = train_test_split(X, y, train_size = 0.8, test_size=0.2, random_state=0)
X_test , X_val , y_test , y_val = train_test_split(X_tmp , y_tmp , train_size=0.5 , test_size=0.5)
```

```python
# Your Code goes here:
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
k_list = np.linspace(1e-2 , 1 , 25)
ca = []
predict_proba = []
for k in k_list:
    clf = LogisticRegression(penalty = 'l2' , C= k , max_iter=1000).fit(X_train, y_train)
    pred_proba_X = clf.predict_proba(X_test)
    score = log_loss(y_test , pred_proba_X)
    ca.append(score)
    predict_proba.append(clf.predict_proba(X_test))
    
    print('Performance on test with {} nearest neighbours: {:.3f}'.format(k, score))
    
plt.scatter(k_list , ca)
plt.title('L2 Logistic Regression Log Loss Padj')
plt.savefig('../Plots/L2_LogLoss_padj.png' , dpi  = 300)
```

```python
clf = LogisticRegression(penalty='l2' , C=0.1 , max_iter=1000).fit(X_train,y_train)
sk.metrics.plot_confusion_matrix(clf , X_val , y_val )
plt.title('Padj Accuracy = %2.1f%%' % (accuracy_score(clf.predict(X_val) , y_val)*100))
plt.savefig('../Plots/L2_LogReg_padj_cm.png' , bbox_inches = 'tight' , dpi = 300)
print(accuracy_score(clf.predict(X_val) , y_val))
```
